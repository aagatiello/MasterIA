**Ejercicio 1**

Supongamos que tenemos datos con cuatro características sobre películas que nos gustaría clasificar en una de tres clases: comedia romántica, terror, documental. ¿Cuál de las siguientes dos arquitecturas de redes neuronales es más poderosa (“más poderosa” significa que puede hacer todo lo que la otra puede hacer y más)?
1. Una sola red neuronal con 4 entradas, 15 neuronas en la capa oculta y 3 neuronas en la capa de salida.
2. Tres redes neuronales separadas, cada una con 4 entradas, 5 neuronas en la capa oculta y 1 neurona
en la capa de salida.

**RESPUESTA**
La primera opción es la más potente porque con una sola red neuronal se pueden recibir las cuatro características de entrada y en la capa de salida hay una neurona especializada por cada clase. La segunda opción serviría para entrenar cada clase por separado y no es la más óptima.
